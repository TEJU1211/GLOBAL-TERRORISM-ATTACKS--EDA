The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.</b>

## DATA CLEANING AND PREPARATRION
TO DO :

*  CHECKING THE CSV FILE

```python
#INFORMATION ON terrorism_df
terrorism_df.info()
```

```python
terrorism_df.columns
```

```python
# CREATING DATAFRAME FOR SELECTIVE COLUMNS 
terrorism_df = terrorism_df[['iyear', 'imonth', 'iday','country_txt','region_txt','provstate', 'city', 'latitude', 'longitude','summary' ,'attacktype1_txt','targtype1_txt', 'gname','motive',
                             'weaptype1_txt','nkill', 'nwound',]]
terrorism_df                            

           
                            
```

```python
# RENAMING THE COLUMNS
terrorism_df.rename(columns = {'iyear':'Year', 'imonth':'Month', 'iday':'Day','country_txt':'Country','region_txt':'Region', 'attacktype1_txt':'Attack_type','targtype1_txt':'target_type',
                               'weaptype1_txt':'weapon_type','nkill':'Killed', 'nwound':'Wounded'}, inplace= True)
terrorism_df
```

```python
# APPLYING DF.INFO TO SEE THE CHANGES
terrorism_df.info()
```

```python
# CHECKING FOR MISSING VALUES
terrorism_df.isna().sum().sort_values(ascending= False)
```

```python
#Filling the nan values in state column as unknown
terrorism_df['provstate']=terrorism_df['provstate'].fillna('Unknown')
```

```python
#Filling the nan values in city column as unknown
terrorism_df['city']=terrorism_df['city'].fillna('Unknown')
```

```python
#Filling the nan values in killed column as 0
terrorism_df['Killed']=terrorism_df['Killed'].fillna(0)
```

```python
#Filling the nan values in killed column as 0
terrorism_df['Wounded']=terrorism_df['Wounded'].fillna(0)
```

```python
#Filling the nan values of Latitude to 0
terrorism_df['latitude'] = terrorism_df['latitude'].fillna(0)
```

```python
#Filling the nan values of Longitude to 0
terrorism_df['longitude'] = terrorism_df['longitude'].fillna(0)
```

```python
#Filling the nan values in summary column as unknown
terrorism_df['summary']=terrorism_df['summary'].fillna('Unknown')
```

```python
#filling the nan values in motive as unknown 
terrorism_df['motive']=terrorism_df['motive'].fillna('Unknown')
```

```python
#Filling the nan values in city as unknown
terrorism_df['city'] = terrorism_df['city'].fillna('Unknown')
```

```python
terrorism_df["Casualty"]=terrorism_df["Killed"]+terrorism_df["Wounded"]
```

```python
#PLOTTING LINE GRAPH FOR KILLED AND WOUNDED
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)


KWC[:10].plot(kind='line',color='blue', figsize=(15,7),ax=ax0)
ax0.set_title('PEOPLE KILLED IN COUNTRIES')
ax0.set_xlabel('COUNTRIES')
ax0.set_ylabel('NO OF PEOPLE KILLED')


KWC1[:10].plot(kind='line', color='green',figsize=(15,7),ax=ax1)
ax1.set_title('PEOPLE WOUNDED IN COUNTRIES')
ax1.set_xlabel('COUNTRIES')
ax1.set_ylabel('WOUNDED')
```
